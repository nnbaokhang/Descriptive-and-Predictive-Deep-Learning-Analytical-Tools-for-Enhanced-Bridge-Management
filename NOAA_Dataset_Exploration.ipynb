# Examine zipcodes from NBI Colorado 2019 and see how many bridges are in 80015 and 80221 respectively

```python
df = pd.read_csv("NBI_Colorado_zipcodes.csv")
```

```python
df
```

```python
count = 0
for i, row in df.iterrows():
    if row["0"] == "80015":
        count += 1
        
print("For the year 2019 for Colorado in NBI, there are " + str(count) + " bridges within 80015")
```

```python
count = 0
for i, row in df.iterrows():
    if row["0"] == "80221":
        count += 1
        
print("For the year 2019 for Colorado in NBI, there are " + str(count) + " bridges within 80221")
```
